## Preparando el ambiente y dataset

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

## 1. Analisis de Facturación

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
Total1 = tienda['Precio'].sum()
Total2 = tienda2['Precio'].sum()
Total3 = tienda3['Precio'].sum()
Total4 = tienda4['Precio'].sum()

Ingresos_totales = pd.Series({'Tienda 1': [Total1], 'Tienda 2': [Total2], 'Tienda 3': [Total3], 'Tienda 4': [Total4]})
Ingresos_totales
```

## 2. Ventas por categoría


En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
Ventas_categoria1 = tienda['Categoría del Producto'].value_counts()
Ventas_categoria2 = tienda2['Categoría del Producto'].value_counts()
Ventas_categoria3 = tienda3['Categoría del Producto'].value_counts()
Ventas_categoria4 = tienda4['Categoría del Producto'].value_counts()
Ventas_categoria = pd.DataFrame({'Tienda 1': Ventas_categoria1, 'Tienda 2': Ventas_categoria2, 'Tienda 3': Ventas_categoria3, 'Tienda 4': Ventas_categoria4})
Ventas_categoria.sort_values(by='Tienda 1', ascending=False)
```

## 3. Calificación promedio de la tienda

En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
AvgTienda = tienda['Calificación'].mean()
AvgTienda2 = tienda2['Calificación'].mean()
AvgTienda3 = tienda3['Calificación'].mean()
AvgTienda4 = tienda4['Calificación'].mean()

print(f'La calificación promedio de la tienda 1 es: {AvgTienda:.2f}')
print(f'La calificación promedio de la tienda 2 es: {AvgTienda2:.2f}')
print(f'La calificación promedio de la tienda 3 es: {AvgTienda3:.2f}')
print(f'La calificación promedio de la tienda 4 es: {AvgTienda4:.2f}')
```

## 4. Productos más y menos vendidos


En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
print('Top 5 productos más vendidos de la tienda 1')
top1 = tienda["Producto"].value_counts().head()
top1
```

```python
print('Top 5 productos menos vendidos de la tienda 1')
tail1 = tienda["Producto"].value_counts().tail().sort_values()
tail1
```

```python
print('Top 5 productos más vendidos de la tienda 2')
top2= tienda2["Producto"].value_counts().head()
top2
```

```python
print('Top 5 productos menos vendidos de la tienda 2')
tail2 = tienda2["Producto"].value_counts().tail().sort_values()
tail2
```

```python
print('Top 5 productos más vendidos de la tienda 3')
top3 = tienda3["Producto"].value_counts().head()
top3
```

```python
print('Top 5 productos menos vendidos de la tienda 3')
tail3 = tienda3["Producto"].value_counts().tail().sort_values()
tail3
```

```python
print('Top 5 productos más vendidos de la tienda 4')
top4 = tienda4["Producto"].value_counts().head()
top4
```

```python
print('Top 5 productos menos vendidos de la tienda 4')
tail4 = tienda4["Producto"].value_counts().tail().sort_values()
tail4
```

## 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
print(f'El Costo de envío promedio para la tienda 1 es de: {tienda["Costo de envío"].mean():,.2f}')
print(f'El Costo de envío promedio para la tienda 2 es de: {tienda2["Costo de envío"].mean():,.2f}')
print(f'El Costo de envío promedio para la tienda 3 es de: {tienda3["Costo de envío"].mean():,.2f}')
print(f'El Costo de envío promedio para la tienda 4 es de: {tienda4["Costo de envío"].mean():,.2f}')
```

## Informe Final

El presente análisis nace de la necesidad que tiene el Sr. Juan por determinar cuál de las 4 tiendas de su cadena Alura Store debe vender para iniciar un nuevo emprendimiento.

Para ello analizaremos los datos más relevantes para ayudar al Sr. Juan a tomar esa decisión.

### 1. Facturación Total

De acuerdo con los datos proporcionados, se puede apreciar que la tienda Nº 1 fue la que mayor ingreso generó y la tienda Nº 4 fue la de menor ingreso.

```python
Facturacion_total = [Total1, Total2, Total3, Total4]
Tiendas_totales = ['Tienda 1','Tienda 2','Tienda 3', 'Tienda 4']

plt.figure(figsize=(10, 6))
bars = plt.bar(Tiendas_totales, Facturacion_total)
plt.title('Facturación Total por tienda')
plt.ylabel('Facturación en miles de millones COP')
plt.xticks(rotation=45)
for bar in bars:
    total = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, total, f"${total:,.0f}", ha='center', va='bottom', fontsize=8)
plt.show()

```

### 2. Productos vendidos por categoría

De acuerdo a las catergorías de productos más vendidos se puede observar que las categorías de "Muebles" y "Electrónicos" encabezan las ventas en todas las tiendas.

```python
Ventas_categoria.sort_values(by='Tienda 1').plot(kind='barh', stacked=False, figsize=(10, 6), width=0.8)
plt.title('Cantidad de productos vendidos por categoría')
plt.grid(axis='x', linestyle='--')
plt.show()

```

### 3. Calificación promedio de la tienda

Al observar la calificación promedio de cada tienda, podemos apreciar que la tienda Nº 3 fue la de mayor puntaje con 4.05, mientras que la de menor puntaje fue la tienda Nº 1 con 3.98

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], [AvgTienda, AvgTienda2, AvgTienda3, AvgTienda4], marker='o', linestyle='-')
plt.title('Calificación promedio de la tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación promedio 1')
plt.grid(axis='y', linestyle='--')
plt.show()
```
